```python
## function
def processText(tableEl):
    holder = []
    if (len(tableEl) >= 14):
        holder = [tableEl[5].text,tableEl[9].text,tableEl[13].text]
    else :
        holder = ['Nan','Nan','Nan']
    if (len(tableEl)>=24):
        holder.append(tableEl[23].text)
    else :
        holder.append('Nan')
    return holder
```

```python
## Main script
YearMonthStore = []
for year in range(2006,2016):
    for month in range(1,13):
        url = 'https://www.wunderground.com/history/monthly/KPHL/date/'+str(year) + '-' + str(month) + '-' + str(1)
        browser.get(url)
        time.sleep(5)
        element = browser.find_elements_by_class_name("summary-table")
        if len(element) > 0:
            tableEl = element[0].find_elements_by_css_selector("td")
            holder = processText(tableEl)
            YearMonthStore.append(holder + [year,month])
        else :
            YearMonthStore.append(['Missing'] * 4 + [year,month])

        
        
        
```

```python
weatherData = pd.DataFrame(YearMonthStore)
weatherData.columns = ['Max','Average','Min','Rain','Year','Month']
weatherData.to_csv('weatherPhiladelphiaData20062015.csv')
```
